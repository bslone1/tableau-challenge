```python
#import dependencies
import pandas as pd
```

```python
#read in all CVS files and add month and year columns to each.

june2019 = pd.read_csv("Resources/201906-citibike-tripdata.csv")
june2019["month"] = "June"
june2019["year"] = 2019
june2019.head()
```

```python
july2019 = pd.read_csv("Resources/201907-citibike-tripdata.csv")
july2019["month"] = "July"
july2019["year"] = 2019
july2019.head()
```

```python
august2019 = pd.read_csv("Resources/201908-citibike-tripdata.csv")
august2019["month"] = "August"
august2019["year"] = 2019
august2019.head()
```

```python
june2020 = pd.read_csv("Resources/202006-citibike-tripdata.csv")
june2020["month"] = "June"
june2020["year"] = 2020
june2020.head()
```

```python
july2020 = pd.read_csv("Resources/202007-citibike-tripdata.csv")
july2020["month"] = "July"
july2020["year"] = 2020
july2020.head()
```

```python
august2020 = pd.read_csv("Resources/202008-citibike-tripdata.csv")
august2020["month"] = "August"
august2020["year"] = 2020
august2020.head()
```

```python
#merge and concatenate all of the above dataframes into one big dataframe
summer_merger = [june2019, july2019, august2019, june2020, july2020, august2020]

summer_tripdata = pd.concat(summer_merger)

summer_tripdata.head(10)
```

```python
#turn new dataframe into csv and send to resource folder for use in tableau (this might take a bit)
summer_tripdata.to_csv("Resources/citibike_summer_2019_2020.csv")
```
